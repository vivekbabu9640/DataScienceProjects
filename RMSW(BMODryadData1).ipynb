```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("/content/BMO+Dryad+Data.xlsx")
```

```python
df.head(5)
```

```python
df.columns = df.iloc[0]
```

```python
df = df[1:]
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.describe()
```

```python
#converting to numeric columns
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
#stress score changes visualization
if set(["pss_-_before_course", "pss_-_course_completed", "pss_-_one_month_on"]).issubset(df.columns):
  df[["pss_-_before_course", "pss_-_course_completed", "pss_-_one_month_on"]].plot(kind="box")
  plt.title("Distribution of Stress Scores Across Time Points")
  plt.ylabel("PSS Score")
  plt.show()
```
